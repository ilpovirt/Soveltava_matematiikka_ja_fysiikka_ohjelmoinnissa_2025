```python
#Vapaa pudotus
import numpy as np
import matplotlib.pyplot as plt
g = 9.81
def vapaa_pudotus(m,k,t):
    v = m*g/k*(1-np.exp(-k/m*t))
    return v
#Aika
t = np.arange(0,60,0.01)
#Kokeillaan pudusta erilaisilla kappaleilla, joilla vaihteleva massa ja muoto.
#Muoto määrittelee vakion k
plt.plot(t,vapaa_pudotus(1,1,t), label = "m = 1kg, k = 1")
plt.plot(t,vapaa_pudotus(5,0.8,t), label = "m = 5kg, k = 0.8")
plt.plot(t,vapaa_pudotus(5,0.6,t), label = "m = 5kg, k = 0.6")
plt.plot(t,vapaa_pudotus(10,2,t), label = "m = 10kg, k = 2")
plt.grid()
plt.legend()
plt.ylabel('Nopeus [m/s]')
plt.xlabel('Aika [s]')
plt.show()
```

```python
#jäähtymislaki
def jaahtuminen(Ty,T0,k,t):
    T = Ty + (T0-Ty)*np.exp(-k*t)
    return T

t = np.arange(0,10,0.01)
plt.plot(t,jaahtuminen(10,100,1,t), label = "Ty = 10,  T0 = 100, k = 1")
plt.grid()
plt.legend()
plt.ylabel('Lämpötila')
plt.xlabel('Aika [s]')
plt.show()
```
