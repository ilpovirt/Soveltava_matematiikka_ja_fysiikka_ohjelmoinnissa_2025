```python
#Tehtv 5, piirretn sinifunktion kuvaaja:

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,20,0.01) #Mritn x:n arvot, np.arange(pienin arvo, suurin arvo, pisteiden vlimata) 
f = 3 * np.sin(2 * x + 1)
f2 = 3 * np.cos(2 * x + 1 - np.pi/2) #Kosinifunktio
plt.figure(figsize=(10,3)) #Mritelln kuvaajan koko
plt.plot(x,f,'b.') #Piirretn kuvaaja
plt.plot(x,f2,'r') 
plt.legend(['Sini','Kosini'])
plt.grid()
plt.show()
```
