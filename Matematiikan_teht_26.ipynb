```python
#Tehtävä 26: 
## Analyyttiset ratkaisut SymPyllä:

import sympy as sm
from sympy.abc import x

#Määritellään, että ratkaistava funktio on nimeltään y
y = sm.Function('y')
# a) y' + 7y = 0
# Tässä y' kirjoitetaan sm.Derivative(y(x),x)

#Ratkaisu a-kohtaan
y_ratk_a = sm.dsolve(sm.Derivative(y(x),x) + 7*y(x),y(x)) #Pilkun oikealla puolella y(x) määrittelee mitä ratkaistaan
print('Yhtälön ratkaisu tehtävässä 26 a on:')
display(y_ratk_a)

#26 b) y' + cos(x)y = 0
y_ratk_b = sm.dsolve(sm.Derivative(y(x),x) + sm.cos(x)*y(x),y(x))
print('Yhtälön ratkaisu tehtävässä 26 b on:')
display(y_ratk_b)

#26 c) xy' + 3x^2y = 0
y_ratk_c = sm.dsolve(x*sm.Derivative(y(x),x) + 3*x**2*y(x) ,y(x))
print('Yhtälön ratkaisu tehtävässä 26 c on:')
display(y_ratk_c)
```

```python
#Numeerinen ratkaisu
#Analyyttisiä ratkaisuja (kaavaa) ei ole olemassa kaikille differentiaaliyhtälöille.
#Joskus se voi olla myös erittäin vaikea löytää.

#Numeerinen ratkaisu ei anna kaavaa, vaan y:n arvot x:n funktiona, joista voi vaikka piirtää
#ratkaisun kuvaajan:

#Esim. b-kohta
#y' + cos(x)y = 0
#Kirjoitetaan eri muodossa:
#y' = -cos(x)y
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc

#Esitetään y' funktiona Pythonissa
def fun1(y,x):
    dydx = - np.cos(x)*y
    return dydx

#Määritellään x:n arvojoukko, eli väli, jolla ratkaisua haetaan
x = np.arange(0,4,0.1)

#Alkuehto (välttämätön numeerisissa ratkaisuissa)
y0 = 2 # y(0) = 2, y:n arvo x:n ensimmäiselle arvolle

ratkaisu = sc.integrate.odeint(fun1,y0,x)

plt.plot(x,ratkaisu)
#Verrataan tunnettuun analyyttiseen ratkaisuun
#Vakio C pitäää ensin ratkaista alkuehdosta
C = 2
plt.plot(x,C*np.exp(-np.sin(x)),'r.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
