```python
#Yhtln ratkaisu y = 4/3x^3 + C
def ratkaisu(x,C):
    return 4/3*x**3+C

import matplotlib.pyplot as plt
import numpy as np
#C:n arvoja
C_arvot = [-2,-1,0,1,2,6]
x = np.arange(-3.3,2,0.1)

plt.figure(figsize=(6,6))
for C in C_arvot:
    plt.plot(x,ratkaisu(x,C))
plt.grid()
plt.legend(C_arvot)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Yhtln ratkaisu y = 4/3x^3 + C
def ratkaisu(x,C):
    return -3/(C + 4*x**3)

import matplotlib.pyplot as plt
import numpy as np
#C:n arvoja
C_arvot = [2]
x = np.arange(0,2,0.1)

plt.figure(figsize=(6,6))
for C in C_arvot:
    plt.plot(x,ratkaisu(x,C))
plt.grid()
plt.legend(C_arvot)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
