```python
# 2. Suorita kiihtyvyyden mittaus, jonka aikana liikuttelet puhelinta eri suuntiin. 
# Tee liike systemaattisesti, esimerkiksi ensin noin 5s ylös-alas (puhelimen suunnassa), 
# sitten 5s oikealle-vasemmalle ja lopuksi 5s eteen-taakse. 

# a) Tallenna havainnot csv-tiedostoon ja tuo se tietokoneelle Pandas-kirjastoa käyttäen. 
import pandas as pd
#Luetaan data Pandas DataFrameen
df = pd.read_csv('./Data/Kiihtyvyys_05112025.csv')
df.head()
```

```python
# b) Piirrä havaintojen perusteella matplotlib.pyplot-kirjastoa käyttäen kuva, 
# jossa kiihtyvyyden kolme komponenttia ja kokonaiskiihtyvyys on esitetty eri väreillä.
#c) Lisää kuvaaan legend, joka kertoo minkä värinen kuvaaja vastaa mitäkin komponenttia.
#d) Lisää kuvaan akselien nimet, otsikko ja gridiviivat. Rajaa akselit sopivasti

import matplotlib.pyplot as plt #Ladataan kirjasto
plt.figure(figsize=(14,4))
plt.plot(df['Time (s)'],df['Linear Acceleration x (m/s^2)'], label = 'Acceleration x')
plt.plot(df['Time (s)'],df['Linear Acceleration y (m/s^2)'], label = 'Acceleration y')
plt.plot(df['Time (s)'],df['Linear Acceleration z (m/s^2)'], label = 'Acceleration z')
plt.plot(df['Time (s)'],df['Absolute acceleration (m/s^2)'], label = 'Acceleration tot')
plt.title('Kiihtyvyysmittaus')
plt.xlabel('Aika [s]')
plt.ylabel('Kiihtyvyys [m/$s^2$]')
plt.grid()
plt.axis([0,17,-10,10]) #[x_min,x_max, y_min,y_max]
plt.legend()
plt.show()
```

```python
# f)Laske kaikkien kolmen komponentin ja kokonaiskiihtyvyyden keskiarvot ja keskihajonta koko mittauksen ajalta.

#Keskiarvo
print('Datan sarakkeiden keskiarvot ovat:')
display(df.mean())

#Keskihajonta
print('Datan sarakkeiden keskihajonnat ovat:')
display(df.std())
```
